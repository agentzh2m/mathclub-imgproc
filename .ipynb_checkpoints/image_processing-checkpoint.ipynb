## Basic Image Processing with Python 2

```python
from matplotlib import pyplot as plt
%matplotlib inline
import numpy as np
```

### Image One

```python
img1 = plt.imread('images/ice.jpg')
plt.imshow(img1)
```

```python
def filter_ice(img):
    new_img = np.copy(img)
    new_img[:,:,0] += 5 
    new_img[:,:,1] /= 2
    new_img[:,:,2] += 5
    return new_img


plt.imshow(filter_ice(img1))
            
```

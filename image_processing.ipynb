## Basic Image Processing with Python 2

```python
from matplotlib import pyplot as plt
%matplotlib inline
import numpy as np
from scipy import ndimage
import scipy
from scipy import misc
```

### Image One

```python
img1 = plt.imread('images/ice.jpg')
plt.imshow(img1)
```

```python
def filter_ice(img):
    new_img = np.copy(img)
    new_img[:,:,0] += 5 
    new_img[:,:,1] /= 2
    new_img[:,:,2] += 3
    return new_img


plt.imshow(filter_ice(img1))
            
```

```python
img2 = plt.imread('images/sea.jpg')
plt.imshow(img2)
```

```python
#Write your code here
new_img2 = np.copy(img2)
plt.imshow(ndimage.gaussian_filter(new_img2, sigma=6))
```

### Image Three

```python
img3 = plt.imread('images/landscape.jpg')
plt.imshow(img3)
```

```python
new_img3 = ndimage.imread('images/landscape.jpg')

for i, space in enumerate(new_img3):
    for j, pixel in enumerate(space):
        for k, c in enumerate(pixel):
            if c < 20:
                new_img3[i][j][k] /= 2
            if c > 200:
                new_img3[i][j][k] = 255
                
plt.imshow(new_img3)
```

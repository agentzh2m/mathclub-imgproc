## Basic Image Processing with Python 2

```python
from matplotlib import pyplot as plt
%matplotlib inline
import numpy as np
from scipy import ndimage
import scipy
from scipy import misc
```

### Image One

```python
img1 = plt.imread('images/ice.jpg')
plt.imshow(img1)
```

```python
def filter_ice(img):
    new_img = np.copy(img)
    new_img[:,:,0] += 5 
    new_img[:,:,1] /= 2
    new_img[:,:,2] += 3
    return new_img


plt.imshow(filter_ice(img1))
            
```

### Image Three

```python
img3 = plt.imread('images/landscape.jpg')
plt.imshow(img3)
```

```python
new_img3 = ndimage.imread('images/landscape.jpg')

for i, space in enumerate(new_img3):
    for j, pixel in enumerate(space):
        for k, c in enumerate(pixel):
            if c < 20:
                new_img3[i][j][k] /= 2
            if c > 200:
                new_img3[i][j][k] = 255
                
plt.imshow(new_img3)
```

```python
new_im6 = np.copy(im6)
for i, space in enumerate(new_im6):
    for j, pixel in enumerate(space):
        for k,c in enumerate(pixel):
            if c > 10:
                new_im6[i][j][k] = np.random.randint(200, 255)
                
plt.imshow(new_im6)
```

## Image 7

```python
im7 = plt.imread('images/muji.jpg')
plt.imshow(im7)
```

```python
new_im7 = np.copy(im7)

for i, space in enumerate(new_im7):
    for j, pixel in enumerate(space):
        for k, c in enumerate(pixel):
            tmp = new_im7[i][j][k]
            low = np.clip(tmp-30, 0, 255)
            high = np.clip(tmp+30, 0, 255)
            new_im7[i][j][k] = np.random.randint(low, high)
                
plt.imshow(new_im7)
```

## Image 8

```python
im8 = plt.imread('images/google.png')
plt.imshow(im8)
```

```python
new_im8 = np.copy(im8)
for i, space in enumerate(new_im8):
    for j, pixel in enumerate(space):
        if np.all(pixel == np.array([1., 1.,  1.,  1.])):
            new_im8[i, j] = np.array([0., 0., 0., 1.])
        elif np.all(pixel >=  np.array([0.9, 0.2, 0.2, 1.])) or np.all(pixel <  np.array([0.93, 0.23, 0.23, 1.])):
            new_im8[i, j] = np.array([0.5, .5, .5, .5])
            
            
plt.imshow(new_im8)
```
